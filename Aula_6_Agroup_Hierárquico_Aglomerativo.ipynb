```
# Importação das bibliotecas e métodos necessários
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler 
```

# Ex 1

```
# Gera um conjunto de 200 pontos com coordenadas aleatórias, divididos em 4 grupos (clusters)
from sklearn.datasets import make_blobs
Atrib, Rot = make_blobs(n_samples = 200,n_features = 2,centers = 4,cluster_std = 1.8,random_state = 42)
 
plt.figure(figsize = (10,10))
plt.scatter(Atrib[:,0],Atrib[:,1],c = Rot)
plt.show()
```

```
# Pré-processamento (Padronização dos dados)
X = Atrib

escalonador = StandardScaler()
escalonador.fit(X)
X_norm = escalonador.transform(X)
```

```
# Agrupamento Hierárquico Aglomerativo com 2 clusters
modeloCLH = AgglomerativeClustering()
modeloCLH.fit(X_norm)
```

```
# Exibição as informações geradas pelo clusterizador
print(f'Total de clusters gerados: {modeloCLH.n_clusters_}')
print(f'Rótulos: {modeloCLH.labels_}')
```

```
# Exibição do resultado
plt.figure(figsize = (10,10))
plt.scatter(Atrib[:,0],Atrib[:,1],c = modeloCLH.labels_)
plt.show()
```

```
# Agrupamento Hierárquico Aglomerativo com 3 clusters
modeloCLH3 = AgglomerativeClustering(n_clusters=3)
modeloCLH3.fit(X_norm)
```

```
# Exibição as informações geradas pelo clusterizador
print(f'Total de clusters gerados: {modeloCLH3.n_clusters_}')
print(f'Rótulos: {modeloCLH3.labels_}')
```

```
# Exibição do resultado
plt.figure(figsize = (10,10))
plt.scatter(Atrib[:,0],Atrib[:,1],c = modeloCLH3.labels_)
plt.show()
```

```
# Agrupamento Hierárquico Aglomerativo com 4 clusters
modeloCLH4 = AgglomerativeClustering(n_clusters=4)
modeloCLH4.fit(X_norm)
```

```
# Exibição as informações geradas pelo clusterizador
print(f'Total de clusters gerados: {modeloCLH4.n_clusters_}')
print(f'Rótulos: {modeloCLH4.labels_}')
```

```
# Exibição do resultado
plt.figure(figsize = (10,10))
plt.scatter(Atrib[:,0],Atrib[:,1],c = modeloCLH4.labels_)
plt.show()
```

```
# Exibição as informações geradas pelo clusterizador
print(f'Total de clusters gerados: {modeloCLH5.n_clusters_}')
print(f'Rótulos: {modeloCLH5.labels_}')
```

```
# Exibição do resultado
plt.figure(figsize = (10,10))
plt.scatter(Atrib[:,0],Atrib[:,1],c = modeloCLH5.labels_)
plt.show()
```

```
# Traçado do dendrograma
from scipy.cluster.hierarchy import dendrogram, linkage

plt.figure(figsize = (10,10))
dendro = dendrogram(linkage(X_norm, method = 'ward'))
plt.title('Dendrograma')
plt.xlabel('Objeto')
plt.ylabel('Distância Euclidiana')
plt.show()
```
